```python
# Import libraries
import pandas as pd
import numpy as np
import re
import time
import json
from selenium import webdriver
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
```

```python
driver_path = '/Users/itselios/Downloads/chromedriver-mac-x64/chromedriver'

# Set up Chrome options (optional)
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--disable-extensions')  # Optional, but recommended

# Create the WebDriver instance
driver = webdriver.Chrome(executable_path=driver_path, options=chrome_options)

# Now you can proceed with your web scraping code
url = 'http://magical-menagerie.com/wizardry/full-character-listing/'
driver.get(url)

# Get elements
elements = driver.find_element(By.CLASS_NAME, 'postcontent')
sub_elem = elements.find_elements(By.TAG_NAME, 'p')
```

```python
# Extract tags which holds names
# tags = [(element.find_elements(By.TAG_NAME, 'strong'), element.find_elements(By.TAG_NAME, 'img')) for element in sub_elem]
tags = []
img_tags = []

for element in sub_elem:
    img_elems = []
    tag = element.find_elements(By.TAG_NAME, 'strong')
    tags.append(tag)
    if (tag == 'Aesalon, Falcon'): break
    img_elems = element.find_elements(By.TAG_NAME, 'img')
    img_tags_inner = []
    for elem in img_elems:
        el = elem.get_attribute('alt')
        if el == "huffle": el = "hufflepuff" 
        if el == "slyth": el = "slytherin" 
        if el == "slug": el = "slug club" 
        if el == "raven": el = "ravenclaw" 
        if el == "quidd": el = "quidditch" 
        if el == "gryff": el = "gryffindor"        
        img_tags_inner.append(el)
    img_tags.append(img_tags_inner)

print(len(tags))
print(len(img_tags))
print(tags)
print(img_tags)


# # Eleminate empty results
# tags_upd = [elem for elem in tags if elem != []]

# # Extracte names
# names = [{'names':t.text} for tag in tags_upd[1:-1] for t in tag if t.text != '']
```

```python
data = {'names': tags, 'images': img_tags}

names_img = pd.DataFrame(data)
# print(names_img)

for index, row in names_img.iterrows():
    if row['names'] == []:
#     Eleminate empty results
        names_img.drop(index, inplace=True)
    else:
#         Extract names
        tag = row['names']
        if tag != []:
            for t in tag:
                names_img.at[index, 'names'] = t.text
print(names_img)
```
